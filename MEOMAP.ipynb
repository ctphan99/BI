```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from scipy.stats import ttest_ind
```

```python
#####import raw data
df1 = pd.read_csv('t6_10.csv')
df2 = pd.read_csv('t11.csv')
df1.info()
```

```python
#choose t7/t8 data only
df1['Date']= pd.to_datetime(df1['vnOrderDate'])
t78=df1[(df1['Date'] >= '2020-07-01') & (df1['Date'] <= '2020-08-31')]
t78
```

```python
#choose data T7   and      data T8 (ungroup)
t7=df1[(df1['Date'] >= '2020-07-01') & (df1['Date'] <= '2020-07-31')]
t8=df1[(df1['Date'] >= '2020-08-01') & (df1['Date'] <= '2020-08-31')]
t7

#order number by user id of T7 (all user)----general set for MM and nonMM Users in T7
cp_gen7= t7.groupby(['userId','merchantCity']).agg(order_no=('couponCode_0','count')).reset_index()
cp_gen7

#order number by user id of T8 (all user)----general set for MM and nonMM Users in T8
cp_gen8= t8.groupby(['userId','merchantCity']).agg(order_no=('couponCode_0','count')).reset_index()
t8
```

```python
#Meo Map data (ungroup)
# only look at MM User in T8 --> FIND MM USERS
mm = t78.loc[t78['couponCode_0'].str.contains("MEOMAP", case=False, na= False)]
mm8 = mm[(mm['Date'] >= '2020-08-01') & (mm['Date'] <= '2020-08-31')]

# ==> GROUP MM USERS
cp8= mm8.groupby(['userId','merchantCity']).agg(order_no=('couponCode_0','count')).reset_index()
cp8
```

```python
#Order quantity of MM User in T8 for any type of orders (from T8)
cp8_unsorted= cp_gen8.merge(cp8['userId'], on='userId', how='right', suffixes=('_8','')).fillna(0)
print(cp8_unsorted)

#Order quantity of MM User in T7 for any type of orders (from T8)
cp7_unsorted= cp_gen7.merge(cp8['userId'], on='userId', how='right', suffixes=('_7','')).fillna(0)
print(cp7_unsorted)
```

```python
#mean order of MM Users for each city in t7
print(np.mean(cp7_unsorted.order_no[cp7_unsorted.merchantCity=='Hà Nội']))
print(np.mean(cp7_unsorted.order_no[cp7_unsorted.merchantCity=='Hồ Chí Minh']))
#mean order of MM Users for each city in t8
print(np.mean(cp8_unsorted.order_no[cp8_unsorted.merchantCity=='Hà Nội']))
print(np.mean(cp8_unsorted.order_no[cp8_unsorted.merchantCity=='Hồ Chí Minh']))
```

```python
#choose subset of MM User to match with high-order nonMM User in t7
cp7=cp7_unsorted[(cp7_unsorted.order_no>9) & (cp7_unsorted.order_no<15)]
print(cp7)

#choose subset of MM User to match with high-order nonMM User in t8
cp8f=cp8_unsorted[(cp8_unsorted.order_no>13) & (cp8_unsorted.order_no<19)]
print(cp8f)
```

```python
#Order quantity of T7 and T8 FOR MM USER
mm78_change= cp8f.merge(cp7[['userId','order_no']], on='userId', suffixes=('_8','_7')).fillna(0)

mm78_change['Order_Change']=mm78_change['order_no_8']-mm78_change['order_no_7']
mm78_change
```

```python
#non-Meo users change in m7/m8
#remove MM users (ones in T8) from general users
nonmm = pd.concat([t78, mm8, mm8]).drop_duplicates(keep=False)

#non-MM Users data in T8
nonmm8 = nonmm[(nonmm['Date'] >= '2020-08-01') & (nonmm['Date'] <= '2020-08-31')]
#non-MM Users data in T7
nonmm7 = nonmm[(nonmm['Date'] >= '2020-07-01') & (nonmm['Date'] <= '2020-07-31')]
nonmm7
```

```python
#Group Non-MM User in T8 for any type of orders (from MM Users T8)
noncp8_unosrted= nonmm8.groupby(['userId','merchantCity']).agg(order_no=('couponCode_0','count')).reset_index()

#Order quantity of MM User in T7 for any type of orders (from MMUsers T8)
noncp7_unsorted= nonmm7.groupby(['userId','merchantCity']).agg(order_no=('couponCode_0','count')).reset_index()
noncp7_unsorted
noncp7_unsorted.to_csv('nonMMJul')
```

```python
#choose subset of nonMM to match with MM high-order Users
noncp7=noncp7_unsorted[(noncp7_unsorted.order_no>9) & (noncp7_unsorted.order_no<15)]
print(noncp7)
#choose subset of nonMM to match with MM high-order Users
noncp8=noncp8_unosrted[(noncp8_unosrted.order_no>13) & (noncp8_unosrted.order_no<19)]
print(noncp8)
```

```python
#Order quantity of T7 and T8 FOR NON-MM USER
nonmm78_change= noncp8.merge(noncp7[['userId','order_no']], on='userId', suffixes=('_8','_7')).fillna(0)

nonmm78_change['Order_Change']=nonmm78_change['order_no_8']-nonmm78_change['order_no_7']
nonmm78_change
```

```python
##############################################Test the significant difference of MM and NonMM increment in order from t7- t8
mm_hcm = mm78_change[mm78_change.merchantCity == 'Hồ Chí Minh']['Order_Change']
nonmm_hcm = nonmm78_change[nonmm78_change.merchantCity == 'Hồ Chí Minh']['Order_Change']

mm_hcm_mean = np.mean(mm_hcm)
nonmm_hcm_mean = np.mean(nonmm_hcm)
print("HCM Meo Map change mean value:",mm_hcm_mean)
print("HCM nonMM change mean value:", nonmm_hcm_mean)

mm_hcm_std = np.std(mm_hcm)
nonmm_hcm_std = np.std(nonmm_hcm)
print("HCM Meo Map change std value:", mm_hcm_std)
print("HCM nonMM change std value:", nonmm_hcm_std)

ttest_hcm, pval_hcm = ttest_ind(mm_hcm,nonmm_hcm)
print("p-value",pval_hcm)

if pval_hcm <0.005:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```

```python
##############################################Test the significant difference of MM and NonMM increment in order from t7- t8
mm_hn = mm78_change[mm78_change.merchantCity == 'Hà Nội']['Order_Change']
nonmm_hn = nonmm78_change[nonmm78_change.merchantCity == 'Hà Nội']['Order_Change']

mm_hn_mean = np.mean(mm_hn)
nonmm_hn_mean = np.mean(nonmm_hn)
print("HA Noi Meo Map change mean value:",mm_hn_mean)
print("Ha Noi nonMM change mean value:", nonmm_hn_mean)

mm_hn_std = np.std(mm_hn)
nonmm_hn_std = np.std(nonmm_hn)
print("HN MM-Order change std value:", mm_hn_std)
print("HN nonMM-Order change std value:", nonmm_hn_std)

ttest_hn, pval_hn = ttest_ind(mm_hn,nonmm_hn)
print("p-value",pval_hn)

if pval_hn <0.005:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```
